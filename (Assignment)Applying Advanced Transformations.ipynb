### Observed Column Combinations for Superhero Info

```python
superhero_info = pd.read_csv("Data/superhero_info - superhero_info.csv")
superhero_info.info()
superhero_info.head()
```

Columns to separate:
   * Hero|Publisher
   * Measurements
       * Height (numeric)
       * Weight (numeric)

```python
superhero_info['Hero|Publisher'].str.split(pat="|",expand=True)
```

```python
superhero_info[['Hero','Publisher']] = superhero_info['Hero|Publisher'].str.split(pat="|",expand=True)
superhero_info.head()
```

```python
superhero_info = superhero_info.drop(columns=['Hero|Publisher'])
superhero_info.head()
```

```python
hero_column_change = superhero_info.pop("Hero")
pub_column_change = superhero_info.pop("Publisher")

superhero_info.insert(0,"Hero",hero_column_change)
superhero_info.insert(1,"Publisher",pub_column_change)

superhero_info.head()
```

```python
# Convert string column of dictionaries
superhero_info['Measurements'] = superhero_info['Measurements'].str.replace("'",'"')

superhero_info['Measurements'] = superhero_info['Measurements'].apply(json.loads)
superhero_info.head()
```

```python
measurements = superhero_info['Measurements'].apply(pd.Series)
superhero_info = pd.concat((superhero_info, measurements),axis=1)

# Drop Measurements column
superhero_info = superhero_info.drop(columns=["Measurements"])

superhero_info.head()
```

```python
# Make a list of all characters to replace
to_replace = [" cm"," kg"]
# run a loop to replace all of the characters in the list at once

for char in to_replace:
    superhero_info['Height'] = superhero_info['Height'].str.replace(char,'',regex=True)
    superhero_info['Weight'] = superhero_info['Weight'].str.replace(char,'',regex=False)
  
superhero_info.head()
```

### Observed Column Combinations for Superhero Powers

```python
superhero_powers = pd.read_csv("Data/superhero_powers - superhero_powers.csv")
superhero_powers.info()
superhero_powers.head()
```

```python
superhero_powers['Powers_split'] = superhero_powers['Powers'].str.split(pat=',',n=0)
superhero_powers.head()
```

```python
## exploding the column of lists
exploded = superhero_powers.explode('Powers_split')

## saving the unique values from the exploded column
cols_to_make = exploded['Powers_split'].dropna().unique()

for col in cols_to_make:
    superhero_powers[col] = superhero_powers['Powers'].str.contains(col)
superhero_powers.info()
superhero_powers.head()
```
